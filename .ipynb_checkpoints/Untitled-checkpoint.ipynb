```python
# Define the product types and years
product_types = ['Product A', 'Product B', 'Product C']
years = ['2022', '2023', '2024', '2025']

# Generate random data
data = np.random.randint(100, size=(4, 3))

# Create a DataFrame from the data
df = pd.DataFrame(data, index=years, columns=product_types)
df = df.reset_index().rename(columns={'index': 'years'})

# Reshape the DataFrame using melt
df_melt = df.melt('years', var_name='product', value_name='values')
df_melt
```

```python

# Create the Altair chart
chart = alt.Chart(df_melt).mark_bar().encode(
    x='years',
    y='values',
    color='product',
    column='product'
).properties(
    width=600/3  # Set the desired width of the chart
)

# Render the chart
chart.display()
```
